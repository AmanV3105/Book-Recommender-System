```python
df_rat_books = df_rat.merge(df , on="ISBN") 
```

```python
df_rat_books
```

```python
book_total_rating = df_rat_books.groupby("title")["rating"].count().reset_index()
```

```python
book_total_rating
```

```python
book_total_rating.rename(columns={
    "rating" : "total_rating"
} , inplace=True)
```

```python
book_total_rating
```

```python
df_final = df_rat_books.merge(book_total_rating , on="title")
```

```python
df_final.head()
```

```python
df_final = df_final[df_final["total_rating"]>=50]
```

```python
df_final.head()
```

```python
df_final.shape
```

```python
df_final.drop_duplicates(["user_id" , "title"] , inplace=True)
```

```python
df_final
```

```python
df_final.shape
```

```python
rating_pivot = df_final.pivot_table(columns="user_id" , index="title" , values="rating")
```

```python
rating_pivot
```

```python
rating_pivot.shape
```

```python
rating_pivot.fillna(0 , inplace=True)
```

```python
rating_pivot
```

```python
from scipy.sparse import csr_matrix
rating_sparse = csr_matrix(rating_pivot)
```

```python
rating_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
model.fit(rating_sparse)
```

```python
distance , suggestion = model.kneighbors(rating_pivot.iloc[237, :].values.reshape(1, -1) , n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(rating_pivot.index[suggestion[i]])
    
```

```python
import pickle
pickle.dump(model , open("artifacts/model.pkl" , "wb"))
pickle.dump(books_name , open("artifacts/books_name.pkl" , "wb"))
pickle.dump(df_final , open("artifacts/final_rating.pkl" , "wb"))
pickle.dump(rating_pivot , open("artifacts/rating_pivot.pkl" , "wb"))
```

```python
def recommend_book(book_name):
    book_index = np.where(rating_pivot.index == book_name)[0][0]
    distance , suggestion = model.kneighbors(rating_pivot.iloc[book_index, :].values.reshape(1, -1) , n_neighbors=6)
    for i in range(len(suggestion)):
        books = rating_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
recommend_book("A Bend in the Road")
```
